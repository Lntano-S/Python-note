```python
# 2.1
TempStr = eval('input("请输入带有符号的温度值:")')
if TempStr[-1] in ['F','f']:
    C = (eval(TempStr[0:-1]) - 32)/1.8
    print("转换后的温度是{:.0f}C".format(C))
elif TempStr[-1] in ['C','c']:
    F=1.8*eval(TempStr[0:-1])+32
    print("转换后的温度是{:.0f}F".format(F))
else :
    print("输入格式错误")
```

```python
# 2.2
TempStr = input("请输入带有符号的货币:")
if TempStr[-1] in ['元']:
    C = (eval(TempStr[0:-1])/6)
    print("转换后的货币是{:.2f}刀".format(C))
elif TempStr[-1] in ['刀']:
    F = eval(TempStr[0:-1])*6
    print("转换后的货币是{:.2f}元".format(F))
else :
    print("输入格式错误")
```

```python
# 2.3
TempStr = input("请输入带有符号的重量:")
if '千克' in TempStr:
    C = (eval(TempStr[0:-2])*2.2046)
    print("转换后的重量是{:.2f}磅".format(C))
elif TempStr[-1] in ['磅']:
    F = eval(TempStr[0:-1])/2.2046
    print("转换后的重量是{:.2f}千克".format(F))
else :
    print("输入格式错误")
```

```python
#实例2 PythonDraw
import turtle #引入“海龟”库
turtle.setup(650, 350, 200, 200) #设置画布大小和位置
turtle.penup()#抬起画笔，移动时不画线
turtle.fd(-250)# 向后移动250（fd是向前，-250即向后）
turtle.pendown()# 落下画笔，移动时画线
turtle.pensize(25)# 设置画笔尺寸为25
turtle.pencolor("purple")# 设置画笔颜色为紫色
turtle.seth(-40)# 设置海龟的朝向（角度）为-40度
for i in range(4):# 循环4次
    turtle.circle(40, 80)# 画圆弧
    turtle.circle(-40, 80)# 画反向圆弧
turtle.circle(40, 80/2) # 画半段圆弧
turtle.fd(40) # 向前直线移动
turtle.circle(16, 180) # 画半圆
turtle.fd(40 * 2/3) # 向前移动一段距离
```

```python
#2.4
import turtle #引入“海龟”库
turtle.setup(650, 350, 200, 200) #设置画布大小和位置
turtle.penup()#抬起画笔，移动时不画线
turtle.fd(-250)# 向后移动250（fd是向前，-250即向后）
turtle.pendown()# 落下画笔，移动时画线
turtle.pensize(25)# 设置画笔尺寸为25
turtle.seth(-40)# 设置海龟的朝向（角度）为-40度
colors = ['red','orange','yellow','green']
for i in range(4):
    turtle.pencolor(colors[i])# 设置画笔颜色
    turtle.circle(40, 80)# 画圆弧
    turtle.circle(-40, 80)# 画反向圆弧
turtle.circle(40, 80/2) # 画半段圆弧
turtle.fd(40) # 向前直线移动
turtle.circle(16, 180) # 画半圆
turtle.fd(40 * 2/3) # 向前移动一段距离
turtle.done()
```
