```python
#TEST 2
Score = float(input("请输入百分制成绩:"))
if Score >= 90:
    print("优")
elif Score >= 80:
    print("良")
elif Score >= 70:
    print("中")
elif Score >= 60:
    print("及格")
else:
    print("不及格")
```

```python
#TEST 3
mile = float(input("请输入总里程:"))
if mile <= 3:
    print("应付车费:10元")
elif  mile <= 10:
    a = 10 + (mile -3) * 1.2
    print("应付车费:{:.1f}元".format(a))
else:
    b = 18.4 + (mile-10) * 1.5
    print("应付车费:{:.1f}元".format(b))
```

```python
#TEST 4
x = float(input())
if x < 0:
    y = 0
    print("{:}".format(y))
elif  x < 5:
    y = x
    print("{:}".format(y))
elif  x < 10:
    y = 3*x-5
    print("{:}".format(y))
elif  x < 20:
    y = 0.5 * x - 2
    print("{:}".format(y))
else:
    print("0")
```

```python
#TEST 5
year = int(input("请输入一个四位年份:"))
if year % 400 == 0:
    print("该年为闰年")
elif year % 4 == 0 and year % 100 != 0:
    print("该年为闰年")
else:
    print("该年不是闰年")
```
