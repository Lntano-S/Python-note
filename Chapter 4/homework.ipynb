```python
#TEST 8
"""
猜数：计算机产生一个1-10之间的随机数，
用户进行多次猜测并将猜测的数值输入程序
程序对每次输入的数值进行判断，
根据判断结果给出提示，直至猜中为止,
最后打印输出猜对数字的次数。
"""
from random import *
answer = randint(1,10)
n_guess = 1
for i in range(10):
    guess = int(input("请输入一个1-10之间的整数："))
    if guess != answer:
        n_guess += 1
        print("很遗憾，你猜错了，再试一次吧")
    if guess == answer:
        break
print("恭喜你用了{}次猜对了".format(n_guess))
```
