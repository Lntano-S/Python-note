```python
def sum_digits(n: int) -> int:
    # 基准情况：当 n 小于 10 时，直接返回 n
    if n < 10:
        return n
    # 递归情况：返回个位数 + 其余位数的和
    else:
        return n % 10 + sum_digits(n // 10)

print(sum_digits(12345))  
```

```python
def sum(n: int) -> float:
    if n == 0:
        return 0
    else:
        return (1 / n) + sum(n - 1)
    
print(sum(3))
```
