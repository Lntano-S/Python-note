```python
#TEST 1
a = float(input(""))
b = float(input(""))
c = float(input(""))
h = (a + b + c) / 3
if (a+b>c)and(a+c>b)and(b+c>a)and(a-b<c)and(a-c<b)and(b-c<a):
    l = a + b + c
    S = (h*(h-a)*(h-b)*(h-c))**0.5
    print("三角形的周长为{:.1f},面积为{:.1f}".format(l,S))
else:
    print("输入的三边无法构成三角形")
```

```python
#TEST 2
x = float(input("利润："))
if 60 < x <= 100:
    y = 0.01*x
    print("奖金提成为：{:.2f}万元".format(y))
elif 40 < x <= 60:
    y = 0.015*x
    print("奖金提成为：{:.2f}万元".format(y))
elif 30 < x <= 40:
    y = 0.03*x
    print("奖金提成为：{:.2f}万元".format(y))
elif 20 < x <= 30:
    y = 0.05*x
    print("奖金提成为：{:.2f}万元".format(y))
elif 10 < x <= 20:
    y = 0.075*x
    print("奖金提成为：{:.2f}万元".format(y))
elif 0 < x <= 10:
    y = 0.1*x
    print("奖金提成为：{:.2f}万元".format(y))
else:
    print("无奖金")
```

```python
#TEST 3
list_1 = input().split(",")
length = len(list_1)
qualify = 0
unqualify = 0
sum = 0
for i in list_1:
    sum += int(i)
    if int(i) >= 60:
        qualify += 1
    else:
        unqualify += 1
print("{:.2f}".format(sum / length),qualify,unqualify)
```

```python
#TEST 4
for i in range(100,1001):
    a = i // 100
    b = i % 10#处理十位数字时，错用//
    c = i % 10
    num = a**3 + b**3 + c**3
    if i == num:
        print(i)
```

```python
#TEST 5
list = [1,1]
for i in range(18):
    new_num = list[-1] + list[-2]
    list.append(new_num)
for j in range(0, 20):
    if (j + 1) % 5 == 0:
        print(list[j])
    else: 
        print(list[j], end = " ")#得知打印列表元素的方法
```

```python
#TEST 6
n = int(input("请输入一个整数："))
if n <= 1:
    print("{}不是一个素数".format(n))
for i in range(2,n):
    if n % i == 0:
        print("{}是一个素数".format(n))
        break
else:
    print("{}是一个素数".format(n))
```
