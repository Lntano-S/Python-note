```python
#题目5
import math as m
pi = m.pi
r = eval(input("请输入r的值:"))#对于eval()函数 加：就是输出字符串，不加就是执行语句 
h = eval(input("请输入h的值:"))
V = (pi * pow(r, 2) * h) / 3
print(round(V, 2))
```

```python
#题目6
a = "www.moe.gov.cn"
print(a[0])
print(a[0:3])
print(a[-3:]) #去掉右边边界来取到末尾字符
print(len(a))#字符串总长度
print(a.find("o"))#索引具体字符
print(a.count("o"))#统计某字符的出现次数
print(a.replace(".","-"))#字符的替换
print(a.upper())#全部大写
print(a.replace("."," "))#删除标点
```
