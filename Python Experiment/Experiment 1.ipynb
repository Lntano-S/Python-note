```python
#TEST 1
'''
整数序列求和。用户输入任意一个正整数N,
计算1到N(包括1和N)相加之后的结果.
'''
n = int(input("请输入一个整数："))
sum = 0
for i in range(n + 1):
    sum += i + 1
print(f"1到{n}之和为{sum}")
```

```python
#TEST 2 
n = int(input("请输入整数N:")) 
total = 0  
for i in range(1, n + 1):  
    total += i
print(f"1到{n}之和为{total}")
```

```python
#TEST 3：输出九九乘法表
for i in range(1,10):
    for j in range(1,i+1):
        print("{}*{}={}".format(j,i,i*j), end=',')
    print() #内层循环结束后换行
```

```python
#TEST 4：阶乘序列求和
sum, tmp = 0,1
for i in range(1,11):
    tmp *= i
    sum += tmp
print("运算结果是:{}".format(sum))
```

```python
#TEST 6
import random
name = input("请输入名字：")
lucky = 1
random.seed(7)
for c in name:
     lucky +=ord(c)*random.randint(1,77)
print("幸运数字是：{}".format(lucky%777))
```
