```python
def find_max_min(arr):
    if not arr:
        return None,None
    max_val = min_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num
    return max_val,min_val
arr = [2,3,6,9,12,15,8]
result = find_max_min(arr)
print(result)
```

```python
def find_max_min(arr):
    if not arr:
        return None, None
    max_val = min_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num
    return max_val, min_val

# 调用函数并直接打印结果
numbers = [3, 1, 4, 1, 5, 9, 2]
result = find_max_min(numbers)
print(result)  # 输出: (9, 1)
```

```python
def bubble_sort(arr):
    n = len(arr)                           # 第1行：获取数组长度
    for i in range(n):                     # 第2行：外层循环，控制排序轮数
        for j in range(0, n-i-1):          # 第3行：内层循环，处理未排序部分
            if arr[j] > arr[j+1]:          # 第4行：比较相邻元素
                arr[j], arr[j+1] = arr[j+1], arr[j]  # 第5行：交换元素
    return arr                             # 第6行：返回排序后的数组
```
