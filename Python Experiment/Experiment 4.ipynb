```python
#TEST 1
'''
编写三个基础数学函数：
要求：
定义函数 square(n)，计算并返回一个数的平方
定义函数 is_even(n)，判断一个数是否为偶数，返回布尔值
定义函数 max_of_three(a, b, c)，返回三个数中的最大值
'''
def square(n : float) -> float:
    return n ** 2

def is_even(n : bool) -> bool:
    if n % 2 == 0:
        return True
    else:
        return False

def max_of_three(a,b,c : float) -> float:
    return max(a,b,c)

print(square(15))
print(is_even(8))
print(is_even(13))
print(max_of_three(12,16,15))
```

```python
#TEST 3
'''
编写函数 sum(x)，求整数 x 的各位数字之和。
从键盘输入一非负整数，然后调用 sum 函数计算各位数字之和并输出结果
'''

def sum(x : int) -> int:
    x_str = str(x)
    sum = 0
    for i in x_str:
        sum += int(i)
    
    return sum

n = (int(input("请输入一个非负整数：")))
print(f"数字{n}的给为数字相加的和为{sum(n)}")
```

TEST 4
编程计算 $$ \binom{n}{m} = \frac{m!}{n!(m - n)!} $$ 通过定义和调用函数来实现


```python
def fac(x : int) -> int:
    result = 1
    for i in range(1, x + 1):
        result *= i
    return result
        
def comb(n : int, m : int) -> float:
    return fac(m) / (fac(n) * fac(m - n))

n,m = map(int, input().split(","))
print(f"n = {n}, m = {m} 时, C(n,m) = {comb(n,m)}")
```

```python
#TEST 5
'''
编写一个函数 isdiff(n),用来判断参数 n 的各位数字是否互不相同，
若互不相同，则返回 1,否则返回 0,并在主程序中测试该函数
'''
def isdiff(n : int) -> int:
    n_str = str(n)
    length = len(n_str)
    for i in n_str[:length]:
        num_i = int(i)
        for j in n_str[num_i + 1:]:
            if i == j:
                return 0
            else:    
                return 1

print(isdiff(184109))
```

```python
#TEST 6
'''
求 s=a+aa+aaa+……+aaa…aaa (n个a) 的值，其中 a 是 1~9 之间的某个数,n 是一个正整数，
例如当 a=2, n=5 时, s=2+22+222+2222+22222=24690.
要求定义一个函数，该函数根据参数 a 和 n 的值返回表达式 a+aa+aaa+……+aaa…aaa
的值，并在主程序中调用测试该函数
'''
def calculate_sum(a,n):
    s = 0
    for i in range(1, n + 1):
        term = int(str(a) * i)
        s += term
    return s

print(calculate_sum(5,2))    
```
