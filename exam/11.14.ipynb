```python
def encrypt(s):
    result = ""
    for i in str(s):
        new_digit = (int(i) + 5) % 10
        result += str(new_digit)
    
    result_list = list(result)
    if len(result_list) >= 4:
        result_list[0], result_list[3] = result_list[3], result_list[0] 
        result_list[1], result_list[2] = result_list[2], result_list[1]
    
    return ''.join(result_list)

print(encrypt(2345))  
```

```python
def calculator(num1, num2, operator):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "错误：除数不能为 0"
        else:
            return num1 / num2
    else:
        return "错误：无效的运算符"

# 主程序
def main():
    try:
        # 输入两个数字
        num1 = float(input("请输入第一个数字: "))
        num2 = float(input("请输入第二个数字: "))
        operator = input("请输入运算符 (+, -, *, /): ").strip()
        
        # 调用计算器函数
        result = calculator(num1, num2, operator)
        
        # 输出结果
        if isinstance(result, str):  # 如果返回的是错误信息
            print(result)
        else:
            print(f"{num1} {operator} {num2} = {result}")
            
    except ValueError:
        print("错误：请输入有效的数字")
    except Exception as e:
        print(f"发生错误：{e}")

# 运行主程序
if __name__ == "__main__":
    main()

# 也可以直接测试函数
print("测试示例:")
print(f"10 + 5 = {calculator(10, 5, '+')}")
print(f"10 - 3 = {calculator(10, 3, '-')}")
print(f"4 * 6 = {calculator(4, 6, '*')}")
print(f"15 / 3 = {calculator(15, 3, '/')}")
print(f"10 / 0 = {calculator(10, 0, '/')}")
print(f"5 % 2 = {calculator(5, 2, '%')}")
```
