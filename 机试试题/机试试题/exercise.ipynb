```python
文本 = "Hello World"

# 常用操作
长度 = len(文本)              # 11
大写 = 文本.upper()           # HELLO WORLD
小写 = 文本.lower()           # hello world
切片 = 文本[0:5]              # Hello
查找 = 文本.find("World")     # 6
替换 = 文本.replace("H", "h") # hello World
print(查找)
```

```python
列表 = [1, 2, 3, "hello", True]

# 常用操作
列表.append(4)           # 添加元素 [1,2,3,"hello",True,4]
列表.remove(2)           # 删除元素 [1,3,"hello",True,4]
长度 = len(列表)          # 5
切片 = 列表[1:3]          # [3, "hello"]
排序 = sorted([3,1,2])   # [1,2,3]
```

```python
n = int(input("请输入一个整数："))
dict = {}
for i in range(1,n+1):
    dict[i] = i*i
print(dict)
```

```python
binary = input("请输入逗号分隔的4位二进制数: ")
list = binary.split(",")
divisible = []
for i in list:
    num = int(i,2)
    if num % 5 == 0:
        divisible.append(i)
print(",".join(divisible))
```
